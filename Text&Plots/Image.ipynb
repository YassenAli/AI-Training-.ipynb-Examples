```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
m = 3
n = 5

M = np.random.randint(10, size=(m,n))
print(M)

plt.plot(M)
plt.show()
```

```python
plt.imshow(M)
plt.colorbar()
for i in range(m):
    for j in range(n):
        plt.text(j, i, str(M[i,j]), fontsize=20,horizontalalignment='center')
plt.show()
```
