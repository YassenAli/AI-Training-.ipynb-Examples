```python
import matplotlib.pyplot as plt
import numpy as np
from IPython import display
display.set_matplotlib_formats('svg') #Vector based graphic
```

```python
x = np.linspace(.5,5,10)
y1 = np.log(x)
y2 = 2-np.sqrt(x)

plt.plot(x,y1,'bo-',label='Log')
plt.plot(x,y2,'rs-',label='Sqrt')
plt.legend()
plt.show()
```
