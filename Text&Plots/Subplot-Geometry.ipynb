```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.subplot(1,2,1)
plt.plot(np.random.randn(10))

plt.subplot(1,2,2)
plt.plot(np.random.randn(10))

plt.show()
```

```python
fig,ax = plt.subplots(2,1)

x = np.arange(10)

ax[0].plot(x, x**2, 'b')
ax[1].plot(x, np.sqrt(x), 'r')

plt.show()
```

```python
fig,ax = plt.subplots(2,2)

x = np.arange(10)

ax[0][0].plot(x,x)
ax[0][1].plot(x,np.sqrt(x))
ax[1][0].plot(x,x**2)
ax[1][1].plot(x,x**3)

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(3,3, figsize=(9,9))
print(ax.shape)
print(ax.flatten().shape)
for i in ax.flatten(): 
    i.plot(np.random.randn(3,3)) # -> (X,Y) -> (0,3)
```
